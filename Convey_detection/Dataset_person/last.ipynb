```python
import os
import shutil
from pathlib import Path

def merge_datasets(source_paths, destination_path):
    # Create destination directories
    dest_images = Path(destination_path) / 'images'
    dest_labels = Path(destination_path) / 'labels'
    dest_images.mkdir(parents=True, exist_ok=True)
    dest_labels.mkdir(parents=True, exist_ok=True)

    # Sets to keep track of copied files
    copied_images = set()
    copied_labels = set()

    for source_path in source_paths:
        train_path = Path(source_path) / 'train'
        
        # Process images
        image_path = train_path / 'images'
        if image_path.exists():
            for img in image_path.glob('*'):
                if img.name not in copied_images:
                    shutil.copy2(img, dest_images)
                    copied_images.add(img.name)
        
        # Process labels
        label_path = train_path / 'labels'
        if label_path.exists():
            for label in label_path.glob('*'):
                if label.name not in copied_labels:
                    shutil.copy2(label, dest_labels)
                    copied_labels.add(label.name)

    print(f"Merged datasets into {destination_path}")
    print(f"Total unique images copied: {len(copied_images)}")
    print(f"Total unique labels copied: {len(copied_labels)}")

# Define source and destination paths
source_paths = [
    '/Users/oakky/Documents/Nong_Crane/nonsense/Top-view-trained-1',
    '/Users/oakky/Documents/Nong_Crane/nonsense/Top-view-trained-2',
    '/Users/oakky/Documents/Nong_Crane/nonsense/Top-view-trained-3'
]
destination_path = '/Users/oakky/Documents/Nong_Crane/nonsense/dataset_real_final'

# Run the merging process
merge_datasets(source_paths, destination_path)
```

```python
import os
from pathlib import Path

def remove_empty_files(labels_path, images_path):
    labels_dir = Path(labels_path)
    images_dir = Path(images_path)

    removed_files = 0

    for label_file in labels_dir.glob('*.txt'):
        # Check if label file is empty
        if label_file.stat().st_size == 0:
            image_file = images_dir / label_file.with_suffix('.jpg').name

            # Remove empty label file
            label_file.unlink()
            removed_files += 1

            # Remove corresponding image file if it exists
            if image_file.exists():
                image_file.unlink()
                removed_files += 1

            print(f"Removed empty label and corresponding image: {label_file.name}")

    print(f"Removal complete. Total files removed: {removed_files}")
    print(f"Labels removed: {removed_files // 2}")
    print(f"Images removed: {removed_files // 2}")

# Paths to your dataset
labels_path = '/Users/oakky/Documents/Nong_Crane/nonsense/dataset_real_final/labels'
images_path = '/Users/oakky/Documents/Nong_Crane/nonsense/dataset_real_final/images'

# Run the removal process
remove_empty_files(labels_path, images_path)
```

```python
import os
from pathlib import Path

def reclassify_labels(labels_path):
    labels_dir = Path(labels_path)
    modified_files = 0

    for label_file in labels_dir.glob('*.txt'):
        with open(label_file, 'r') as f:
            lines = f.readlines()

        modified = False
        new_lines = []
        for line in lines:
            parts = line.strip().split()
            if len(parts) >= 5 and parts[0] != '0':
                parts[0] = '0'  # Change class to 0
                modified = True
            new_line = ' '.join(parts) + '\n'
            new_lines.append(new_line)

        if modified:
            with open(label_file, 'w') as f:
                f.writelines(new_lines)
            modified_files += 1
            print(f"Modified: {label_file.name}")

    print(f"Reclassification complete. Total files modified: {modified_files}")
 
# Path to your labels
labels_path = '/Users/oakky/Documents/Nong_Crane/nonsense/dataset_real_final/labels'

# Run the reclassification process
reclassify_labels(labels_path)
```

```python
import os
import random
import shutil

def split_train_val_no_overlap(source_folder, output_folder, val_ratio=0.2):
    """
    แบ่งข้อมูลเป็น train และ validation sets โดยไม่มีรูปภาพซ้ำกัน
    
    :param source_folder: โฟลเดอร์ต้นทางที่มีโฟลเดอร์ย่อย 'images' และ 'labels'
    :param output_folder: โฟลเดอร์ปลายทางสำหรับเก็บ train และ val sets
    :param val_ratio: สัดส่วนของ validation set (default: 0.2)
    """
    # สร้างโฟลเดอร์ปลายทาง
    train_images_folder = os.path.join(output_folder, "train", "images")
    train_labels_folder = os.path.join(output_folder, "train", "labels")
    val_images_folder = os.path.join(output_folder, "val", "images")
    val_labels_folder = os.path.join(output_folder, "val", "labels")
    
    os.makedirs(train_images_folder, exist_ok=True)
    os.makedirs(train_labels_folder, exist_ok=True)
    os.makedirs(val_images_folder, exist_ok=True)
    os.makedirs(val_labels_folder, exist_ok=True)

    # รวบรวมรายชื่อไฟล์ภาพและ label ที่ตรงกัน
    image_files = [f for f in os.listdir(os.path.join(source_folder, "images")) 
                   if f.lower().endswith((".jpg", ".jpeg", ".png"))]
    
    valid_pairs = []
    for img_file in image_files:
        base_name = os.path.splitext(img_file)[0]
        label_file = base_name + ".txt"
        if os.path.exists(os.path.join(source_folder, "labels", label_file)):
            valid_pairs.append((img_file, label_file))

    # สุ่มลำดับของคู่ไฟล์
    random.shuffle(valid_pairs)

    # แบ่งข้อมูล
    split_index = int(len(valid_pairs) * val_ratio)
    val_pairs = valid_pairs[:split_index]
    train_pairs = valid_pairs[split_index:]

    # คัดลอกไฟล์ไปยังโฟลเดอร์ปลายทาง
    def copy_files(pairs, img_dest, label_dest):
        for img_file, label_file in pairs:
            shutil.copy(os.path.join(source_folder, "images", img_file), os.path.join(img_dest, img_file))
            shutil.copy(os.path.join(source_folder, "labels", label_file), os.path.join(label_dest, label_file))

    copy_files(train_pairs, train_images_folder, train_labels_folder)
    copy_files(val_pairs, val_images_folder, val_labels_folder)

    print(f"การแบ่งข้อมูลเสร็จสิ้น:")
    print(f"จำนวนภาพทั้งหมด: {len(valid_pairs)}")
    print(f"จำนวนภาพใน training set: {len(train_pairs)}")
    print(f"จำนวนภาพใน validation set: {len(val_pairs)}")

# ตัวอย่างการใช้งาน
split_train_val_no_overlap('/Users/oakky/Documents/Nong_Crane/nonsense/dataset_real_final', '/Users/oakky/Documents/Nong_Crane/nonsense/dataset_real_final_split', val_ratio=0.2)
```
